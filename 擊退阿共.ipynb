```python
def 拋物線(速度,仰角):
    t=symbols('t')
    仰角=(仰角*np.pi)/180
    vy0=速度*np.sin(仰角)
    vx0=速度*np.cos(仰角)
    vyt = vy0 - 9.8*t
    yt = integrate(vyt,t)
    xt = integrate(vx0,t)
    p=50
    q=0
    y=[]
    x=[]
    c=0
    a=10
    while c < a:
        
        y.append(yt.evalf(subs={t:c})-yt.evalf(subs={t:0}))
        
        x.append(xt.evalf(subs={t:c})-xt.evalf(subs={t:0}))
        if (xt-p)**2 + (yt-q)**2 < 25:
            break
        elif yt < 0:
            break
        else:
            c= c+0.01   
    
    plt.axis([0, 60, 0, 100])
    plt.plot(x,y)
    plt.scatter(p,q)
    print(np.sin(仰角),速度)
```

```python
interact_manual(拋物線,速度=(0,10.),仰角=(0.,90));
```
