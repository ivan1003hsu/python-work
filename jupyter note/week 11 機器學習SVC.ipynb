### np.meshgrid

```python
x_list = np.linspace(3,10,50)
y_list = np.linspace(6,8,50)
```

### np.meshgrid(x_list,y_list)

```python
X,Y = np.meshgrid(x_list,y_list)
P= np.c_[X.ravel(),Y.ravel()]
C=model.predict(P)
plt.scatter(P[:,0],P[:,1],c=C)
```
