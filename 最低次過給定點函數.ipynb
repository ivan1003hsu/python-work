```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
from sympy.abc import x,y
from sympy import S
```

```python
n =1
dots=[]
while n!= 0:
    dot = input('請輸入第{}個點座標'.format(n))
    if 'end' in dot:
        n=n-1
        break
    dot = dot.split(',')
    dots.extend(dot)
    n = n+1
dots = np.array(list(map(float,dots)))
i=0
f=0
while i < 2*n:
    k=0
    a=dots[i+1]
    while k < 2*n:
        if k != i:
            a=(a*(x-dots[k]))/(dots[i]-dots[k])
        k = k+2
    f = f +a
    i = i+2
sp.plot(f)
print('函式為f(x)=',end = '')
print(sp.S(f))
##dots = np.array(dots).reshape(n,2)
##sp.plot(dots[:,0],dots[:,1])
```
